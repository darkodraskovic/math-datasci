## Boxplots

```python
m = 30
n = 6

data = np.zeros((m,n))
for i in range(n):
    data[:,i] = 30 * np.random.standard_normal(m) * (2 * i/(n-1) - 1) + (i + 1)**2
```

### matplotlib

```python
# rows represent observations and columns represent features
plt.boxplot(data)
fig.show()
```

```python
# n observations and m features
plt.boxplot(data.T)
fig.show()
```

### seaborn

```python
sns.boxplot(data=data, orient='v')
plt.show()
```

```python
# pandas dataframe
df = pd.DataFrame(data=data, columns=['zero', 'one', 'two', 'three', 'four', 'five'])
sns.boxplot(data=data, orient='h')
plt.show()
```

### standard_normal, uniform

```python
size = 100
scl = 10
data1 = np.random.standard_normal(size) * scl
data2 = np.random.uniform(size=size) * scl - scl/2
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.boxplot(data1)
ax2.boxplot(data2)

plt.show()
```

## Histogram

```python
n = 1000
# log-normal distribution
data = np.exp(np.random.standard_normal(n) * .5)

bins = 40

fig = plt.figure(figsize=(16,8))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.scatter(x=np.arange(n), y=data, marker='s')
ax2.hist(data, bins=bins, orientation='vertical')

# y is hist bin heights, x is hist bin boundaries; len(x) = len(y) + 1
y, x = np.histogram(data, bins=bins)
# find individual hist bin average val, i.e. bin center; len(xx) = len(y)
xx = (x[:-1] + x[1:]) / 2
ax3.plot(xx, y)

plt.show()
```

```python
# turn hist counts into proportions
plt.plot(xx, y / np.sum(y))
```
