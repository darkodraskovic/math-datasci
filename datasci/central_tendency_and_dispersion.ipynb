## Measures of central tendency

```python
from typing import Union

def histogram(d: np.ndarray, nbins: int=30) -> tuple[np.ndarray, np.ndarray]:
    y,x = np.histogram(a=d, bins=nbins)
    x = (x[:-1] + x[1:]) / 2
    return y,x

def create_figure(subplot: int = None) -> Union[plt.Figure, mpl.axes.SubplotBase]:
    fig: plt.Figure = plt.figure(figsize=(8, 6))
    fig.add_axes((0.0, 0.0, 1.0, 1.0))
    if subplot is None: 
        return fig, None
    return fig, fig.add_subplot(subplot)
```

```python
N = 1001
nbins = 30
rng = np.random.default_rng()
```

```python
d1 = rng.standard_normal(N) - 1
d2 = rng.standard_normal(N) * 3
d3 = rng.standard_normal(N) + 1

y1,x1 = histogram(d1, nbins)
y2,x2 = histogram(d2, nbins)
y3,x3 = histogram(d3, nbins)

fig, ax = create_figure(111)

ax.plot(x1,y1, 'r')
ax.plot(x2,y2, 'g')
ax.plot(x3,y3, 'b')

ax.set_xlabel('Data values')
ax.set_ylabel('Data counts')

fig.show()
```
