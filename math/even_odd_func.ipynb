```python
import math
import sympy as sym
import sympy.plotting.plot as symplot

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import display, Math
sym.init_printing()
```

```python
x = np.linspace(-5, 5, 20)

fEven = x**2
fEvenNeg = (-x)**2

plt.plot(x, fEven)
plt.plot(x, fEvenNeg, 'ro')
plt.show()
```

```python
fOdd = x**3
fOddNeg = (-x)**3
# fNegOdd = -x**3
fNegOdd = -fOdd

plt.plot(x, fOdd)
plt.plot(x, fOddNeg, 'ro')
plt.plot(x, fNegOdd, 'g')
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 64)

fSin = np.sin(x)
fSinNeg = np.sin(-x)

plt.plot(x, fSin)
plt.plot(x, fSinNeg)
plt.title("$sin(x)$")
plt.show()
```

```python
fCos = np.cos(x)
fCosNeg = np.cos(-x)
fNegCos = -fCos


plt.plot(x, fCos)
plt.plot(x, fCosNeg, 'ro')
plt.plot(x, fNegCos, 'g')
plt.title("$cos(x)$")
plt.show()
```
